```python
import pandas as pd;
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python
sera_data=pd.read_csv("D:\\Projects\\MoonLight-Energy-Solutions\\Data\\sierraleone.csv")
sera_data.info()
sera_data.describe()
```

```python
sera_data.head()
```

```python
sera_data['Timestamp'] = pd.to_datetime(sera_data['Timestamp'],errors='coerce')
```

```python
sera_data = sera_data.sort_values(by='Timestamp').reset_index(drop=True)
```

```python
sera_data['Timestamp'].head()
sera_data['Timestamp'].isna().sum()
```

```python
sera_data.isna().sum()
```

```python
sera_data.loc[sera_data['GHI'] < 0, 'GHI'] = None
sera_data.loc[sera_data['DNI'] > 1200, 'DNI'] = None
sera_data.loc[sera_data['GHI'] > 1500, 'GHI'] = None
sera_data.loc[sera_data['DNI'] < 0, 'DNI'] = None
sera_data.loc[sera_data['DHI'] < 0, 'DHI'] = None
```

```python
sera_data['Hour']=sera_data['Timestamp'].dt.hour
sera_data['Day']=sera_data['Timestamp'].dt.day
sera_data['Month']=sera_data['Timestamp'].dt.month
sera_data['Year']=sera_data['Timestamp'].dt.year
```

```python
sera_data[['Timestamp','Hour','Day','Month','Year']].head()
```

```python
sera_data['Timestamp'] = pd.to_datetime(sera_data['Timestamp'], errors='coerce')
df = sera_data.set_index('Timestamp')
```

```python
df_daily = sera_data.resample('D', on='Timestamp').sum()
# Convert Wh/m² to kWh/m²
df_daily['GHI_kWh'] = df_daily['GHI'] / 1000
df_daily['DNI_kWh'] = df_daily['DNI'] / 1000
df_daily['DHI_kWh'] = df_daily['DHI'] / 1000
```

```python
df_monthly = sera_data.resample('ME', on='Timestamp').mean()
```

```python
# Diffusion Fraction: This tells us how cloudy or hazy the atmosphere is
sera_data['Diffuse_Fraction'] = sera_data['DHI'] / sera_data['GHI']
```

```python
sera_data['Diffuse_Fraction'].describe()
```

```python
#Direct Beam Strength
sera_data['Beam_Ratio'] = sera_data['DNI'] / sera_data['GHI']
```

```python
# Step 1: Convert the time column to datetime if it's not already
sera_data['Timestamp'] = pd.to_datetime(sera_data['Timestamp'])

# Step 2: Set it as the index
sera_data.set_index('Timestamp', inplace=True)

# Step 3: Now resample and compute standard deviation
ghi_hourly_std = sera_data.resample('H').GHI.std()
```

```python
ghi_daily_std = sera_data.resample('D').GHI.std()
```

```python
df_daily['PSH'] = df_daily['GHI_kWh']
df_daily['PSH'].describe()
```

```python
df_daily.groupby(df_daily.index.month)['PSH'].mean()
```

```python
sera_data[['GHI', 'TModA']].corr()
```

```python
sera_data[['WS', 'TModA']].corr()
```

```python
summary = {
    'Annual_GHI_kWh' : df_daily['GHI_kWh'].sum(),
    'Average_PSH' : df_daily['GHI_kWh'].mean(),
    'Diffuse_Fraction_mean' : sera_data['Diffuse_Fraction'].mean(),
    'Beam_Ratio_mean' : sera_data['Beam_Ratio'].mean(),
    'GHI_hourly_std_mean' : ghi_hourly_std.mean(),
    'GHI_daily_std_mean' : ghi_daily_std.mean(),
    'Avg_Temperature' : df['TModA'].mean(),
    'Avg_Wind' : df['WS'].mean(),
    'Total_Precip_mm' : df['Precipitation'].sum(),
}
```

```python
summary_df=pd.DataFrame([summary])
```

```python
import pandas as pd;
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python
togo_data=pd.read_csv("D:\\Projects\\MoonLight-Energy-Solutions\\Data\\togo.csv")
togo_data.head()
togo_data.info()
togo_data.describe()
```

```python
togo_data['Timestamp'] = pd.to_datetime(togo_data['Timestamp'],errors='coerce')
```

```python
togo_data = togo_data.sort_values(by='Timestamp').reset_index(drop=True)
```

```python
togo_data['Timestamp'].head()
togo_data['Timestamp'].isna().sum()
```

```python
togo_data.isna().sum()
```

```python
togo_data.loc[togo_data['GHI'] < 0, 'GHI'] = None
togo_data.loc[togo_data['DNI'] > 1200, 'DNI'] = None
togo_data.loc[togo_data['GHI'] > 1500, 'GHI'] = None
togo_data.loc[togo_data['DNI'] < 0, 'DNI'] = None
togo_data.loc[togo_data['DHI'] < 0, 'DHI'] = None
```

```python
togo_data['Hour']=togo_data['Timestamp'].dt.hour
togo_data['Day']=togo_data['Timestamp'].dt.day
togo_data['Month']=togo_data['Timestamp'].dt.month
togo_data['Year']=togo_data['Timestamp'].dt.year
```

```python
togo_data[['Timestamp','Hour','Day','Month','Year']].head()
```

```python
togo_data['Timestamp'] = pd.to_datetime(togo_data['Timestamp'], errors='coerce')
df = togo_data.set_index('Timestamp')
```

```python
df_daily = togo_data.resample('D', on='Timestamp').sum()
# Convert Wh/m² to kWh/m²
df_daily['GHI_kWh'] = df_daily['GHI'] / 1000
df_daily['DNI_kWh'] = df_daily['DNI'] / 1000
df_daily['DHI_kWh'] = df_daily['DHI'] / 1000
```

```python
df_monthly = togo_data.resample('ME', on='Timestamp').mean()
```

```python
# Diffusion Fraction: This tells us how cloudy or hazy the atmosphere is
togo_data['Diffuse_Fraction'] = togo_data['DHI'] / togo_data['GHI']
```

```python
togo_data['Diffuse_Fraction'].describe()
```

```python
#Direct Beam Strength
togo_data['Beam_Ratio'] = togo_data['DNI'] / togo_data['GHI']
```

```python
# Step 1: Convert the time column to datetime if it's not already
togo_data['Timestamp'] = pd.to_datetime(togo_data['Timestamp'])

# Step 2: Set it as the index
togo_data.set_index('Timestamp', inplace=True)

# Step 3: Now resample and compute standard deviation
ghi_hourly_std = togo_data.resample('H').GHI.std()
```

```python
ghi_daily_std = togo_data.resample('D').GHI.std()
```

```python
df_daily['PSH'] = df_daily['GHI_kWh']
df_daily['PSH'].describe()
```

```python
df_daily.groupby(df_daily.index.month)['PSH'].mean()
```

```python
togo_data[['GHI', 'TModA']].corr()
```

```python
togo_data[['WS', 'TModA']].corr()
```

```python
# Remove night time and bad sensor readings
df_togo = togo_data[togo_data['GHI']>20].copy()
```

```python
# Keep only valid daylight rows
df_togo_day = togo_data[togo_data['GHI'] > 50].copy()  # GHI > 50 to avoid tiny denominator

# Compute ratios safely
df_togo_day['Diffuse_Fraction'] = df_togo_day['DHI'] / df_togo_day['GHI']
df_togo_day['Beam_Ratio'] = df_togo_day['DNI'] / df_togo_day['GHI']

# Remove impossible values
df_togo_day.loc[(df_togo_day['Diffuse_Fraction'] < 0) | (df_togo_day['Diffuse_Fraction'] > 1), 'Diffuse_Fraction'] = None
df_togo_day.loc[(df_togo_day['Beam_Ratio'] < 0) | (df_togo_day['Beam_Ratio'] > 1.2), 'Beam_Ratio'] = None

# Interpolate missing values
df_togo_day[['Diffuse_Fraction', 'Beam_Ratio']] = df_togo_day[['Diffuse_Fraction', 'Beam_Ratio']].interpolate()

# Check results
df_togo_day[['Diffuse_Fraction','Beam_Ratio']].describe()
```

```python
summary = {
    'Annual_GHI_kWh' : df_daily['GHI_kWh'].sum(),
    'Average_PSH' : df_daily['GHI_kWh'].mean(),
    'Diffuse_Fraction_mean' : df_togo_day['Diffuse_Fraction'].mean(),
    'Beam_Ratio_mean' : df_togo_day['Beam_Ratio'].mean(),
    'GHI_hourly_std_mean' : ghi_hourly_std.mean(),
    'GHI_daily_std_mean' : ghi_daily_std.mean(),
    'Avg_Temperature' : df['TModA'].mean(),
    'Avg_Wind' : df['WS'].mean(),
    'Total_Precip_mm' : df['Precipitation'].sum(),
}
```

```python
summary_df=pd.DataFrame([summary])
```
